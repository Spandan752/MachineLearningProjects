```python
numeric_features = [feature for feature in df.columns if df[feature].dtypes != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtypes == 'O']

#print columns
print("Numeric Features:", numeric_features)
print("Categorical Features:", categorical_features)
```

```python
df['Total_Marks'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['Average'] = df['Total_Marks']/3
df.head(2)
```

Data Visualization

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
plt.subplot(1,2,1)
sns.histplot(data=df, x='Average', kde=True, bins=30, color='red')
plt.subplot(1,2,2)
sns.histplot(data=df, x='Average', kde=True, hue='gender')
plt.show()
```

```python

fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df,x='Total_Marks',bins=30,kde=True,color='blue')
plt.subplot(122)
sns.histplot(data=df,x='Total_Marks',kde=True,hue='gender')
plt.show()
```
